# 5. Clusterização

## 5.1 Normalização dos dados 

```python
normalizador = Normalizer()
df_normalizado = normalizador.fit_transform(df)
```

## 5.2 Gausian Mixture

### 5.2.1 Método de Elbow (cotovelo)

```python
elbow = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df_normalizado)
    elbow.append(kmeans.inertia_)
plt.plot(range(1, 11), elbow, 'bx-')
plt.title('Método Elbow')
plt.xlabel('Número de Cluster')



plt.show()
```

### 5.2.2 Método Silhueta

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score,v_measure_score

alvo = df_normalizado
max = 15
km_scores= []
km_silhouette = []
vmeasure_score =[]
db_score = []
gm_bic= []
gm_score=[]
for i in range(2,max):
    km = KMeans(n_clusters=i, random_state=0).fit(alvo)
    preds = km.predict(alvo)
    
    km_scores.append(-km.score(alvo))
    
    silhouette = silhouette_score(alvo,preds)
    km_silhouette.append(silhouette)

# Plotagem do método da Silhouette (Silhueta)
plt.figure(figsize=(7,4))
plt.title("Silhouette Method",fontsize=16)
plt.scatter(x=[i for i in range(2,max)],y=km_silhouette,s=150,edgecolor='k')
plt.grid(True)
plt.xlabel("Number of clusters",fontsize=14)
plt.ylabel("Silhouette score",fontsize=15)
plt.xticks([i for i in range(2,max)],fontsize=14)
plt.yticks(fontsize=15)
plt.show()
```

 Percebe-se que 2 a inclinação já começa a ser menor, logo a escolha vai ser de 2 clusters

### 5.2.3 Clusterização gaussiana

```python
GM = GaussianMixture(4).fit(df_normalizado)
labels_gaussian_mixture = GM.predict(df_normalizado)
df_cluster=pd.DataFrame(df_normalizado, columns = df.columns)
df_cluster['clusters']=labels_gaussian_mixture.astype('str')
```

# 6. Regressão

```python
df_reg = df_cluster.copy() #copiando o df clusterizado
```

```python
df_reg.head()
```

## 6.1 Fazendo o one-hot-enconding dos clusters

```python
df_reg = pd.get_dummies(df_reg, columns=['clusters'])
```

```python
df_reg.head()
```

## 6.2 Separando Dataset em treino e teste

```python
X = df_reg.drop(columns = ['CREDIT_LIMIT'])
Y = df_reg[['CREDIT_LIMIT']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42, shuffle=True)
```

# 7. Rede Neural MLPRegressor

```python
regr = MLPRegressor().fit(X_train, y_train)
```

```python
y_pred_train = regr.predict(X_train)
y_pred_test = regr.predict(X_test)
```

```python
print("RMSE treino = {}".format(mean_squared_error(y_train, y_pred_train)))
print("R² treino = {}".format(r2_score(y_train, y_pred_train)))
print("MAE treino = {}".format(mean_absolute_error(y_train, y_pred_train)))
```

```python
print("RMSE teste = {}".format(mean_squared_error(y_test, y_pred_test)))
print("R² teste = {}".format(r2_score(y_test, y_pred_test)))
print("MAE teste = {}".format(mean_absolute_error(y_test, y_pred_test)))
```

## 7.1 Otimização de parametros

```python
params_grid = {
    'hidden_layer_sizes' : [(100,), (1,), (50,)],
    'activation' : ['identity', 'logistic', 'tanh', 'relu'],
    'solver' : ['lbfgs', 'sgd', 'adam'],
    'alpha' : [0.0005, 0.0001, 0.01],
    'learning_rate' : ['constant', 'invscaling', 'adaptive']
}
```

```python
clf = MLPRegressor()

grid_search = GridSearchCV(clf, param_grid=params_grid, cv=5, n_jobs=-1, verbose=2)  #cv = cross validation, n_jobs: -1 utiliza todos nucleos da maquin, verbose = quao verboso você precida ver
grid_search.fit(X_train, y_train)
```

```python
y_pred = grid_search.best_estimator_.predict(X_test)
```

```python
print("RMSE = {}".format(mean_squared_error(y_test, y_pred, squared=False)))
print("R² = {}".format(r2_score(y_test, y_pred)))
print("MAE = {}".format(mean_absolute_error(y_test, y_pred)))
```

Conclue-se que a clusterização foi essencial para um melhor aprendizado por parte do modelo, obtendo um R² de 0.99 e MAE de 0.009.
